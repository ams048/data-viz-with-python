```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

start_year = '1980'
end_year = '2010'
df = df[(df['year'] >= start_year) & (df['year'] <= end_year)]

df.head()
```

```python
df.info()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here
#Number of books published per year.
#Number of books published by each publisher.
#Text ratings count versus reviews count.
#Matplotlib

year_counts = df['year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.bar(year_counts.index.astype(int), year_counts.values)
plt.title('Number of Books Published per Year')
plt.xlabel('Year')
plt.ylabel('Number of Books')
plt.show()
```

```python
publisher_counts = df['publisher'].value_counts().head(20)
plt.figure(figsize=(10,5))
plt.bar(publisher_counts.index, publisher_counts.values)
plt.title('Number of Books Published by Each Publisher')
plt.xlabel('Publisher')
plt.ylabel('Number of Books')
plt.xticks(rotation=90)
plt.show()
```

```python
ratings_counts = df['ratings_count']
reviews_counts = df['text_reviews_count']
plt.figure(figsize=(10,5))
plt.bar(['Text Ratings Count', 'Text Reviews Count'], [ratings_counts.sum(), reviews_counts.sum()])
plt.title('Text Ratings Count versus Reviews Count')
plt.ylabel('Count')
plt.show()
```

```python
# Use your Python skills to create a visualization here.
#Number of books published per year.
#Number of books published by each publisher.
#Text ratings count versus reviews count.
#Seaborn

sb.countplot(data=df, x='year', order=df['year'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Books Published per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
sb.countplot(data=df, y='publisher', order=df['publisher'].value_counts().index[:10])
plt.title('Number of Books Published by Each Publisher')
plt.xlabel('Count')
plt.ylabel('Publisher')
plt.show()
```
